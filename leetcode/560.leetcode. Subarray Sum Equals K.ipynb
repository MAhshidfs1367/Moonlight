```python
def subarraySum(nums,k):
        count = 0
        dic = {0:1}
        prefixsum = 0

        for num in nums:
            prefixsum += num
            d = prefixsum - k
            if d in dic:
                count += dic[d]
            if prefixsum in dic:
                dic[prefixsum] += 1
            else:
                dic[prefixsum] = 1
        return count 
```

```python
nums = [1,1,1]
k = 2
subarraySum(nums,k)
```

```python
nums = [1,2,3]
k = 3
subarraySum(nums,k)
```
